```python
import numpy as np

def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

class Perceptron:
    
    def __init__(self, lr=0.1, epochs=20):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = None
    
    def train(self, X, y):
        n_samples, n_features = X.shape
        
        self.weights = np.random.randn(n_features)
        self.bias = np.random.randn()
        
        for _ in range(self.epochs):
            for i in range(n_samples):
                
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = step_function(linear_output)
                
                error = y[i] - y_pred
                
                self.weights = self.weights + self.lr * error * X[i]
                self.bias = self.bias + self.lr * error
    
    def predict(self, X):
        predictions = []
        for x in X:
            linear_output = np.dot(x, self.weights) + self.bias
            predictions.append(step_function(linear_output))
        return np.array(predictions)


X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

AND_y = np.array([0,0,0,1])
OR_y = np.array([0,1,1,1])
NAND_y = np.array([1,1,1,0])
NOR_y = np.array([1,0,0,0])


gates = {
    "AND": AND_y,
    "OR": OR_y,
    "NAND": NAND_y,
    "NOR": NOR_y
}


for gate in gates:
    
    print("\n",gate,"Gate")
    
    model = Perceptron(lr=0.1, epochs=25)
    model.train(X, gates[gate])
    
    predictions = model.predict(X)
    
    for i in range(len(X)):
        print(X[i][0], X[i][1], "->", predictions[i])
```
